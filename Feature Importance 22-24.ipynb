```python
import pandas as pd

df = pd.read_excel('ENADES2022.xlsx')

rename_dict = {
    "p03.1": "equal_opportunity",
    "p03.2": "poverty_individual_blame",
    "p03.4": "upward_mobility_belief",
    "p03.5": "income_inequality_high",
    "p04": "perceived_economic_inequality",
    "p05": "perceived_income_gap_trend",
    "p07": "intergenerational_mobility",
    "p10.1": "inequality_urban_rural",
    "p10.2": "inequality_lima_rest",
    "p10.3": "inequality_rich_poor",
    "p10.4": "inequality_gender",
    "p10.5": "inequality_racial",
    "p10.7": "inequality_lgbtq",
    "p11.1": "inequality_education_access",
    "p11.2": "inequality_health_access",
    "p11.3": "inequality_employment_access",
    "p11.4": "inequality_justice_access",
    "p12.3": "elite_influence",
    "p12.4": "lima_centralism",
    "p17": "progressive_taxation",
}

df = df.rename(columns=rename_dict)

print(df.columns)
```

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np

exclude_cols = ['edadr', 'sexo', 'zona1', 'area2', 'hogar', 'edu2', 'ocupa2', 'NSE2',
               'yhogar', 'etnicidad2', 'ideologia2', 'cluster']

features = [col for col in df.columns if col not in exclude_cols]

df["cluster"] = df["cluster"] + 1

X = df[features]
clusters = df["cluster"].unique()

feature_importance_dict = {}

for c in clusters:
    y = (df["cluster"] == c).astype(int)
    
    model = RandomForestClassifier(
        n_estimators=200,
        random_state=42,
        n_jobs=-1
    )
    
    model.fit(X, y)
    
    feature_importance_dict[c] = model.feature_importances_

fi_df = pd.DataFrame(feature_importance_dict, index=features)

fi_df = fi_df.div(fi_df.sum(axis=0), axis=1)
```

```python
fi_df = fi_df.reindex(sorted(fi_df.columns), axis=1)
```

```python
import matplotlib.pyplot as plt

fi_df["total"] = fi_df.sum(axis=1)
fi_df = fi_df.sort_values("total", ascending=True)
fi_df = fi_df.drop(columns="total")

fi_df = fi_df.reindex(sorted(fi_df.columns), axis=1)

# Plot
fig, ax = plt.subplots(figsize=(10, 8))

fi_df.plot(
    kind="barh",
    stacked=True,
    ax=ax
)

ax.set_title("Feature Importance by Cluster for Social Perception Features - 2022 Survey")
ax.set_xlabel("Relative Importance")
ax.set_ylabel("Features")

ax.legend(title="Cluster")

plt.tight_layout()
plt.show()
```
