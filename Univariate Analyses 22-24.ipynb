```python
import pandas as pd
```

```python
archivo_excel = "ENADES2022.xlsx"
hoja = "BBDD"

df = pd.read_excel(archivo_excel, sheet_name=hoja)
```

```python
df.columns
```

```python
import pandas as pd

df_vars = df.drop(columns=["hogar","p08","dep"], errors="ignore")

resultados = []

for col in df_vars.columns:
    conteo = df_vars[col].value_counts(dropna=False)
    porcentaje = df_vars[col].value_counts(normalize=True, dropna=False) * 100
    
    tabla = pd.DataFrame({
        "Variable": col,
        "Categoría": conteo.index,
        "Conteo": conteo.values,
        "Porcentaje (%)": porcentaje.values.round(1)
    })
    
    tabla = tabla.sort_values(by="Categoría").reset_index(drop=True)
    
    resultados.append(tabla)

resultado_final = pd.concat(resultados, ignore_index=True)

resultado_final.to_excel("resumen_variables_2022.xlsx", index=False)

print("Archivo 'resumen_variables.xlsx' generado correctamente.")
```
